```python
import fitz
def extract_text_from_pdf(pdf_path):
  doc = fitz.open(pdf_path)
  text = ""
  for page in doc:
    text += page.get_text()
  return text
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
  chunk_size = 500,
  chunk_overlap  = 200,
  length_function = len,
)
chunks=text_splitter.split_text(extract_text_from_pdf("/content/budget_speech_pdf_24-25.pdf"))
```

```python
!pip install -U langchain-community
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

embeddings= HuggingFaceEmbeddings()
db=FAISS.from_texts(chunks, embeddings)
```

```python
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
from langchain.chains import RetrievalQA

# Load model and tokenizer from Hugging Face
model_id = "google/flan-t5-base"  # You can replace with another local model
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForSeq2SeqLM.from_pretrained(model_id)

# Wrap the model into a text2text pipeline
hf_pipeline = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=512,
    do_sample=True,
    temperature=0.3,
    top_k=50
)

# Use LangChain's HuggingFacePipeline LLM wrapper
llm = HuggingFacePipeline(pipeline=hf_pipeline)

# Create RAG chain
qa = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=db.as_retriever()
)
```

```python
query = "Summarize the document."
print(qa.run(query))
```
